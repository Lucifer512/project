```python
# removing Stop Words from the string
%pip install nltk
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
Data_in_text = word_tokenize (After_Removing_Pun)
Final_Articles = [word for word in Data_in_text if not word in stopwords.words()]
print(Final_Articles)
```

```python
#importing the K-Means Clustering
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
```

```python
# Applying K-Means Clustering on the String 
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(Final_Articles)
cluster=2
model = KMeans(n_clusters=cluster, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(cluster):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind])
```
